**From the ProPublica notebook**: There are a number of reasons remove rows because of missing data:
* If the charge date of a defendants Compas scored crime was not within 30 days from when the person was arrested, we assume that because of data quality reasons, that we do not have the right offense.
* We coded the recidivist flag -- `is_recid` -- to be -1 if we could not find a compas case at all.
* In a similar vein, ordinary traffic offenses -- those with a `c_charge_degree` of 'O' -- will not result in Jail time are removed (only two of them).
* We filtered the underlying data from Broward county to include only those rows representing people who had either recidivated in two years, or had at least two years outside of a correctional facility.

Now we quantize the dataframe. In particular, we will quantize the priors, the length of stay and the compas score.

```python
dfcutQ = dfcut[['sex','race','age_cat','c_charge_degree','score_text','priors_count','is_recid',
                'two_year_recid','length_of_stay']].copy()

# Quantize priors count between 0, 1-3, and >3
def quantizePrior(x):
    if x <=0:
        return '0'
    elif 1<=x<=3:
        return '1 to 3'
    else:
        return 'More than 3'

    
# Quantize length of stay
def quantizeLOS(x):
    if x<= 7:
        return '<week'
    if 8<x<=93:
        return '<3months'
    else:
        return '>3 months'
    
# Quantize length of stay
def adjustAge(x):
    if x == '25 - 45':
        return '25 to 45'
    else:
        return x

# Quantize score_text to MediumHigh
def quantizeScore(x):
    if (x == 'High')| (x == 'Medium'):
        return 'MediumHigh'
    else:
        return x

    
dfcutQ['priors_count'] = dfcutQ['priors_count'].apply(lambda x: quantizePrior(x))
dfcutQ['length_of_stay'] = dfcutQ['length_of_stay'].apply(lambda x: quantizeLOS(x))
dfcutQ['score_text'] = dfcutQ['score_text'].apply(lambda x: quantizeScore(x))
dfcutQ['age_cat'] = dfcutQ['age_cat'].apply(lambda x: adjustAge(x))
```

Now we set the distortion metric. This function will receive the two dictionary of features X and Y corresponding to the new and old values, and return a distortion value. In this case the distortion function returns 1000 if education is lowered by any value, or increased by more than 1 year, or if age is increased or decreased by more than one decade. In returns a penalty of 1.0 if income is decreased. All other values are 0.

```python
class Dclass():
# adjust education
    def adjustPrior(self,v):
        if v=='0':
            return 0
        elif v=='1 to 3':
            return 1
        else:
            return 2

    def adjustAge(self,a):
        if a == 'Less than 25':
            return 0
        elif a == '25 to 45':
            return 1
        else:
            return 2

    # distortion metric
    def getDistortion(self,vold,vnew):
        '''
        Distortion metric.

        Inputs:
        *vold : dictionary of the form {attr:value} with old values
        *vnew : dictionary of the form {attr:value} with new values

        Output
        *d : distortion value
        '''

        # value that will be returned for events that should not occur
        bad_val = 1e4


        # Adjust prior
        pOld = self.adjustPrior(vold['priors_count'])
        pNew = self.adjustPrior(vnew['priors_count'])

        # Priors cannot be increased, or lowered by more than 1 category. A change has a unit penalty
        if (pNew>pOld)| (pNew<pOld-1):
            return bad_val

        # adjust age
        aOld = self.adjustAge(vold['age_cat'])
        aNew = self.adjustAge(vnew['age_cat'])

        # Age cannot be increased or decreased in more than one category
        if np.abs(aOld-aNew)>1.0:
            return bad_val
        
        # Recidivism should not be increased
        if vold['is_recid'] < vnew['is_recid']:
            return bad_val
        
        cum_sum = 0.0
        
        
        if np.abs(aOld-aNew)>0:
#             cum_sum+=1
#             cum_sum = cum_sum**2
           cum_sum = cum_sum+1
           
        # Penalty of 1 if priors is decreased or increased
        if np.abs(pNew-pOld)>0:
#             cum_sum+=1
#             cum_sum = cum_sum**2
            cum_sum = cum_sum+1
           

           #cum_sum = cum_sum**2
        if vold['is_recid'] > vnew['is_recid']:
#             cum_sum+=1
#             cum_sum = cum_sum**2
            cum_sum = cum_sum+1
           
       

       # final penalty of 2 for changing misdemeanor to felony and vice-verse
        if vold['c_charge_degree'] != vnew['c_charge_degree']:
#             cum_sum+=2
#             cum_sum = cum_sum**2
           cum_sum = cum_sum+4
            
        return cum_sum
        
```

We set the excess distortion constraints (`c1` and `c2`) and set the distortion values. For now, we are solving the problem (to view the equation correctly you have to run the last markdown cell of this notebook with the latex preamble)

\begin{align}
	\min_{p_{\Xh,\Yh|X,Y,D}}& \sum_{x,y} \left| p_{X,Y}(x,y)- p_{\Xh,\Yh}(x,y)\right| \\
	\sto~~&  1-\epsilon\leq \frac{p_{\Yh|D}(y|d_1)}{p_{\Yh|D}(y|d_2)} \leq 1+\epsilon, \forall y\in \calY, d_1,d_2\in \calD\\
	&\mathbb{E}\left( d(x,y,\Xh,\Yh) \mid X=x,Y=y,D=d\right) \leq \delta_D~\forall~(x,y)\in \calX\times \calY,\\
	&p_{\Xh,\Yh|X,Y,D} \mbox{ is a valid distribution.}
\end{align}

We set `c1=.99`, `c2 = 1.99`, and `c3=2.99`.

```python
# c1 = .99 # value of (delta1,c1): to keep.
# c2 = 1.99  # value of (delta2,c2): value that should no happen
# c3 = 2.99 # penalty things that should not happen
# clist = [c1,c2, c3]
Dclass = Dclass()

DT.setDistortion(Dclass)
```

```python
sns.set_style("darkgrid")
sns.set(font_scale=1.8,font='sans-serif')

label_size = 30
plt.rcParams['xtick.labelsize'] = label_size 
plt.rcParams['ytick.labelsize'] = label_size 
plt.rcParams['axes.labelsize'] = label_size 

def getPivot(df, form = 'mean'):
    if form == 'mean':
        dfrecid = (100*df.groupby(['c_charge_degree','age_cat','priors_count'])["is_recid"].mean()).reset_index()
    else:
        dfrecid = (df.groupby(['c_charge_degree','age_cat','priors_count'])["is_recid"].count()).reset_index()
    dfrecidPivot1 = dfrecid.pivot_table(index=['c_charge_degree','age_cat'],columns='priors_count', values="is_recid")
    l = list(dfrecidPivot1.index)
    l.sort(custom_cmp2)       
    dfrecidPivot1 = dfrecidPivot1.loc[l,:]
    return dfrecidPivot1

def plotHeatMapDiff(df1,df2,title='',cut = 20,ax = '',cbar=True,cbar_ax= None,xticklabels=True, yticklabels=True):
    dfrecid = getPivot(df1) 
    dfrecid2 = getPivot(df2)
    
    # mask values that have count below cut -- otherwise results can be exagerated.
    dfMaskPivot = getPivot(df1,form='count')
    dfMaskPivot = dfMaskPivot.applymap(lambda x : np.nan if x<=cut else 0)
    
    dfrecid = dfrecid + dfMaskPivot
    dfrecDiff = dfrecid2-dfrecid
    
    
#     columnOrder = ["<6",6,7,8,9,10,11,12,'>12']
#     dfIncomePivot = dfIncomePivot[columnOrder]
#     t = [str(int(x)) if type(x)!= str else x for x in dfIncomePivot.index ]
#     dfIncomePivot.index = pd.Index(t,name=dfIncomePivot.index.name)
    
    
    sns.set(font_scale=2,font='sans-serif')
    
    if ax == '':
        fig = plt.figure(figsize=(10,7))
        sns.heatmap(dfrecDiff,linewidths=2,annot=True, cbar=cbar,cbar_kws={"label": "Recidivism %"},vmin=-50,vmax=50,xticklabels=xticklabels, yticklabels=yticklabels)
        plt.title(title)
    else:
        if cbar_ax == None:
            sns.heatmap(dfrecDiff,linewidths=2,annot=True,cbar=cbar, cbar_kws={"label": "Recidivism %"},vmin=-50,vmax=50,ax=ax)
            ax.set_title(title,fontsize=30)
        else:
            sns.heatmap(dfrecDiff,linewidths=2,annot=True,cbar=cbar, cbar_ax = cbar_ax, cbar_kws={"label": "Difference in Recidivism %"},vmin=-50,vmax=50,ax=ax)
            ax.set_title(title,fontsize=30)
            
        if not xticklabels:
            ax.get_xaxis().set_ticks([])
        if not yticklabels:
            ax.get_yaxis().set_ticks([])

        ax.set_title(title,fontsize=35)
    
    


def plotPercentage(df1,title='',cut = 20,ax = '',cbar=True,cbar_ax= None,xticklabels=True, yticklabels=True):
    dfrecid = getPivot(df1) 
   
    
    # mask values that have count below cut -- otherwise results can be exagerated.
    dfMaskPivot = getPivot(df1,form='count')
    
    dfMaskPivot = dfMaskPivot.applymap(lambda x : np.nan if x<=cut else 0)
    
    
    dfrecid = dfrecid + dfMaskPivot
    
    sns.set(font_scale=2,font='sans-serif')
    
    if ax == '':
        fig = plt.figure(figsize=(10,7))
        sns.heatmap(dfrecid,linewidths=2,annot=True, cbar=cbar,cbar_kws={"label": "Recidivism %"},vmin=0,vmax=100,xticklabels=xticklabels, yticklabels=yticklabels)
        plt.title(title,fontsize=label_size)
    else:
        if cbar_ax == None:
            sns.heatmap(dfrecid,annot=True,linewidths=2,cbar=cbar, cbar_kws={"label": "Recidivism %"},vmin=0,vmax=100,ax=ax)
        else:
            sns.heatmap(dfrecid,linewidths=2,annot=True,cbar=cbar, cbar_ax = cbar_ax, cbar_kws={"label": "Recidivism %"},vmin=0,vmax=100,ax=ax)
        
        if not xticklabels:
            ax.get_xaxis().set_ticks([])
        if not yticklabels:
            ax.get_yaxis().set_ticks([])

        ax.set_title(title,fontsize=35)
            
    return (dfrecid,dfMaskPivot)
    
    
# Original dataset dataset

# Whole dataset
df = dfcutQ[features].copy()
f, axarr = plt.subplots(2,6, sharey=False,figsize=(42,2*8),gridspec_kw = {'width_ratios':[4,4,4,4,4, .25]})

plotPercentage(df,title='Whole Original Dataset',ax = axarr[0,0],cbar=False)
axarr[0,0].set_xlabel('')
axarr[0,0].set_ylabel('Charged degree - Age')

# AA Female dataset
dfAfricanAmericanF = df.loc[(df['race']=='African-American')&(df['sex']=='Female'),:]
#dfFemale2 = df2.loc[df2['Gender']==' Female',:]
dfrec = plotPercentage(dfAfricanAmericanF,title='African-American Female',ax = axarr[0,1],cbar=False,yticklabels=False)
axarr[0,1].set_xlabel('')
axarr[0,1].set_ylabel('')

# Caucasian Female dataset
dfCaucasianF = df.loc[(df['race']=='Caucasian')&(df['sex']=='Female'),:]
#dfFemale2 = df2.loc[df2['Gender']==' Female',:]
(dfrecid,dfMaskPivot) = plotPercentage(dfCaucasianF,title='Caucasian Female',ax = axarr[0,2],cbar=False,yticklabels=False)
axarr[0,2].set_xlabel('')
axarr[0,2].set_ylabel('')

# AA Male dataset
dfAfricanAmericanM = df.loc[(df['race']=='African-American')&(df['sex']=='Male'),:]
#dfFemale2 = df2.loc[df2['Gender']==' Female',:]
plotPercentage(dfAfricanAmericanM,title='African-American Male',ax = axarr[0,3],cbar=False,yticklabels=False)
axarr[0,3].set_xlabel('')
axarr[0,3].set_ylabel('')

# Caucasian Male dataset
dfCaucasianM = df.loc[(df['race']=='Caucasian')&(df['sex']=='Male'),:]
#dfMinority2 = df2.loc[df2['Race']=='Minority',:]
plotPercentage(dfCaucasianM,title='Caucasian Male',ax = axarr[0,4],yticklabels=False,cbar_ax = axarr[0,5])
axarr[0,4].set_xlabel('')
axarr[0,4].set_ylabel('')


########################
## difference
plotHeatMapDiff(df,df2,title='',ax = axarr[1,0],cbar=False)
axarr[1,0].set_xlabel('Number of priors')
axarr[1,0].set_ylabel('Charged degree - Age')

# AA Female dataset
dfAfricanAmericanF2 = df2.loc[(df2['race']=='African-American')&(df2['sex']=='Female'),:]
#dfFemale2 = df2.loc[df2['Gender']==' Female',:]
plotHeatMapDiff(dfAfricanAmericanF,dfAfricanAmericanF2,title='',ax = axarr[1,1],cbar=False,yticklabels=False)
axarr[1,1].set_xlabel('Number of priors')
axarr[1,1].set_ylabel('')

# Caucasian Female dataset
dfCaucasianF2 = df2.loc[(df2['race']=='Caucasian')&(df2['sex']=='Female'),:]
#dfFemale2 = df2.loc[df2['Gender']==' Female',:]
plotHeatMapDiff(dfCaucasianF,dfCaucasianF2,title='',ax = axarr[1,2],cbar=False,yticklabels=False)
axarr[1,2].set_xlabel('Number of priors')
axarr[1,2].set_ylabel('')

# AA Male dataset
dfAfricanAmericanM2 = df2.loc[(df2['race']=='African-American')&(df2['sex']=='Male'),:]
#dfFemale2 = df2.loc[df2['Gender']==' Female',:]
plotHeatMapDiff(dfAfricanAmericanM,dfAfricanAmericanM2,title='',ax = axarr[1,3],cbar=False,yticklabels=False)
axarr[1,3].set_xlabel('Number of priors')
axarr[1,3].set_ylabel('')

# Caucasian Male dataset
dfCaucasianM2 = df2.loc[(df2['race']=='Caucasian')&(df2['sex']=='Male'),:]
#dfMinority2 = df2.loc[df2['Race']=='Minority',:]
plotHeatMapDiff(dfCaucasianM,dfCaucasianM2,title='Caucasian Male',ax = axarr[1,4],yticklabels=False,cbar_ax = axarr[1,5])
axarr[1,4].set_xlabel('Number of priors')
axarr[1,4].set_ylabel('')



# # Caucasian Female dataset
# dfCaucasianF = df.loc[(df['race']=='Caucasian')&(df['sex']=='Female'),:]
# #dfFemale2 = df2.loc[df2['Gender']==' Female',:]
# (dfrecid,dfMaskPivot) = plotPercentage(dfCaucasianF,title='Caucasian Female',ax = axarr[0,2],cbar=False,yticklabels=False)
# axarr[0,2].set_xlabel('')
# axarr[0,2].set_ylabel('')

# # AA Male dataset
# dfAfricanAmericanM = df.loc[(df['race']=='African-American')&(df['sex']=='Male'),:]
# #dfFemale2 = df2.loc[df2['Gender']==' Female',:]
# plotPercentage(dfAfricanAmericanM,title='African-American Male',ax = axarr[0,3],cbar=False,yticklabels=False)
# axarr[0,3].set_xlabel('')
# axarr[0,3].set_ylabel('')

# # Caucasian Male dataset
# dfCaucasianM = df.loc[(df['race']=='Caucasian')&(df['sex']=='Male'),:]
# #dfMinority2 = df2.loc[df2['Race']=='Minority',:]
# plotPercentage(dfCaucasianM,title='Caucasian Male',ax = axarr[0,4],yticklabels=False,cbar_ax = axarr[0,5])
# axarr[0,4].set_xlabel('')
# axarr[0,4].set_ylabel('')

# df3 = df.copy()
# df3["Income Binary"] = df2["Income Binary"]

# # Whole dataset
# #f, axarr = plt.subplots(1,5, sharey=False,figsize=(48,8),gridspec_kw = {'width_ratios':[4,4,4,4, .25]})

# plotHeatMapDiff(df,df3,title='Difference of Income Percentage',ax = axarr[1,0],cbar=False)

# # Female dataset
# dfFemale1 = df.loc[df['Gender']==' Female',:]
# dfFemale2 = df3.loc[df3['Gender']==' Female',:]
# plotHeatMapDiff(dfFemale1,dfFemale2,title='',ax = axarr[1,1],cbar=False)
# axarr[1,1].set_ylabel('')

# # Male dataset
# dfMale1 = df.loc[df['Gender']==' Male',:]
# dfMale2 = df3.loc[df3['Gender']==' Male',:]
# plotHeatMapDiff(dfMale1,dfMale2,title='',ax = axarr[1,2],cbar=False)
# axarr[1,2].set_ylabel('')

# # Minority dataset
# dfMinority1 = df.loc[df['Race']=='Minority',:]
# dfMinority2 = df3.loc[df3['Race']=='Minority',:]
# plotHeatMapDiff(dfMinority1,dfMinority2,title='',ax = axarr[1,3],cbar_ax = axarr[1,4])
# #axarr[1,3].set_xlabel('')
# axarr[1,3].set_ylabel('')

# plt.suptitle('Adult Dataset Values Before and After Randomization',fontsize=35)#\n$\epsilon=$'+str(epsilon)+', $\delta_1=$'+str(dlist[0])+', $\delta_2=$'+str(dlist[1])+', $\delta_3=$'+str(dlist[2]),fontsize=40)
plt.savefig('./plots/DiffPlot_3.pdf')
```
